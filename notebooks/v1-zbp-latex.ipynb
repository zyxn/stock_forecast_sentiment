```python
import pandas as pd
```

```python
data = pd.read_csv("../data/external/ihsg 14-12-22 till 22-04-24.csv")
ayam = data.head(5)
```

```python
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set gaya untuk seaborn
sns.set(style="whitegrid")

# Grafik Harga Penutupan
plt.figure(figsize=(14, 7))
sns.lineplot(x=data["Date"], y=data['Open'])
plt.title('Harga Pembukaan IHSG')
plt.xlabel('Tanggal')
plt.ylabel('Harga Pembukaan (Rupiah)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('harga_pembukaan_ihsg.png')  # Menyimpan gambar ke komputer
plt.show()

```

```python

# Grafik Volume Perdagangan
plt.figure(figsize=(14, 7))
sns.lineplot(x=data["Date"], y=data['Volume'], color='orange')
plt.title('Volume Perdagangan IHSG')
plt.xlabel('Tanggal')
plt.ylabel('Volume Perdagangan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('volume_perdagangan_ihsg.png')  # Menyimpan gambar ke komputer
plt.show()
```
