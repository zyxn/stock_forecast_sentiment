```python
import torch
from transformers import AutoTokenizer, BertForSequenceClassification
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
text = "Duo of Jumbo Bank Shares Makes JCI Fall Again"

inputs = tokenizer(text, return_tensors="pt")
with torch.no_grad():
    logits = model(**inputs).logits
    
predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```

```python
import pandas as pd

df = pd.read_csv("../data/processed/TransResult.csv")
df
```

```python
column_name = "Judul_english	"  # Ganti "text_column" dengan nama kolom yang Anda inginkan
inputs = df[column_name].tolist()

# Melakukan tokenisasi
tokenized_inputs = tokenizer(inputs, return_tensors="pt", padding=True, truncation=True)

# Memproses input menggunakan model
with torch.no_grad():
    logits = model(**tokenized_inputs).logits

# Memprediksi kelas
predicted_class_ids = logits.argmax(dim=1).tolist()

# Menambahkan hasil prediksi ke DataFrame
print(predicted_class_ids)
```
