```python
df = pd.read_csv("../data/raw/result.csv",delimiter=";",quotechar='"')
df.head(10)
```

```python
import pandas as pd

# Menggunakan fungsi try-except untuk menangani kesalahan saat mengonversi tanggal
for index, row in df.iterrows():
    try:
        df.at[index, 'tanggal'] = pd.to_datetime(row['tanggal'], format="%d %B %Y %H:%M")
    except Exception as e:
        print(f"Error pada baris {index}: {e}")
        # Menghapus baris yang menyebabkan kesalahan
        df.drop(index, inplace=True)

# Menyegarkan ulang indeks DataFrame setelah menghapus baris
df.reset_index(drop=True, inplace=True)
df['tanggal'] = pd.to_datetime(df['tanggal'], format="%d %B %Y %H:%M")
```
